```python
import matplotlib.pylab as plt
import pandas as pd
import numpy as np
import seaborn as sns
import glob
```

```python
version = '6'
data = {}
data['UncSampling'] = {}
data['RandomSampling'] = {}

for i in [0,299]:
    fname = '../../../zenodo/UncSampling/class_prob/v' + version + \
                  '/class_prob_UncSampling__loop_' + str(i) + '.csv'
    
    data['UncSampling'][i] = pd.read_csv(fname)
    
    fname2 = '../../../zenodo/RandomSampling/class_prob/v' + version + \
                  '/class_prob_RandomSampling_ini_cond_US__loop_' + str(i) + '.csv'
    
    data['RandomSampling'][i] = pd.read_csv(fname2)
```

```python
sns.set(rc={'axes.facecolor': 'white', 'figure.facecolor': 'white'})
c = ['#F5622E', '#15284F']

fig = plt.figure(figsize=(8,5))

plt.subplot(1,1,1)
sns.distplot(data['RandomSampling'][0]['prob_Ia'][data['RandomSampling'][0]['type'] == 'Ia'], 
             kde=True, label='Ia', bins=100, color=c[1])
sns.distplot(data['RandomSampling'][0]['prob_Ia'][data['RandomSampling'][0]['type'] != 'Ia'], 
             kde=True, label='nonIa', bins=100, color=c[0])
plt.legend(title='Initial state')
plt.xlabel('prob_Ia',fontsize=18)
plt.ylabel('PDF', fontsize=18)
plt.axvline(0.5, ls='--')

plt.show()
#plt.savefig('../../plots/prob_evolution_loop0.pdf')
```

```python

fig = plt.figure(figsize=(16, 5))

plt.subplot(1,2,1)
sns.distplot(data['RandomSampling'][299]['prob_Ia'][data['RandomSampling'][299]['type'] == 'Ia'], 
             kde=True, label='Ia -- loop 300', bins=100, color=c[1])
sns.distplot(data['RandomSampling'][299]['prob_Ia'][data['RandomSampling'][299]['type'] != 'Ia'], 
             kde=True, label='nonIa -- loop 300', bins=100, color=c[0])
plt.legend(title='RandomSampling')
plt.xlabel('prob_Ia',fontsize=18)
plt.ylabel('PDF', fontsize=18)
plt.axvline(0.5, ls='--')

plt.subplot(1,2,2)
sns.distplot(data['UncSampling'][299]['prob_Ia'][data['UncSampling'][299]['type'] == 'Ia'], 
             kde=True, label='Ia -- loop 300', bins=100, color=c[1])
sns.distplot(data['UncSampling'][299]['prob_Ia'][data['UncSampling'][299]['type'] != 'Ia'], 
             kde=True, label='nonIa -- loop 300', bins=100, color=c[0])
plt.legend(title='UncSampling')
plt.xlabel('prob_Ia',fontsize=18)
plt.ylabel('PDF', fontsize=18)
plt.axvline(0.5, ls='--')

plt.subplots_adjust(hspace=0.3)
plt.show()
#plt.savefig('../../plots/prob_evolution_loop299.pdf')
```
